```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_iris
```

```
data, target = load_iris(return_X_y=True)
```

```
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.2, stratify=target, random_state=14300631)
```

```
model = LogisticRegression()
model.fit(X_train, y_train)
```

```
y_pred_train = model.predict(X_train)
```

```
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
```

```
print(classification_report(y_train, y_pred_train))
```

```
y_pred_test = model.predict(X_test)
```

```
print(classification_report(y_test, y_pred_test))
```

```
df = pd.read_csv('mushrooms.csv')
```

```
df.head()
```

```
df['class'] = df['class'].replace(to_replace='p', value=1)
df['class'] = df['class'].replace(to_replace='e', value=0)
```

```
target = df['class']
X = df.drop('class', axis=1)
```

```
X_dummies = pd.get_dummies(X)
```

```
X_train, X_test, y_train, y_test = train_test_split(X_dummies, target, stratify=target)
```

```
model = LogisticRegression()
model.fit(X_train, y_train)
```

```
y_predict_test = model.predict(X_test)
print(classification_report(y_test, y_predict_test))
```
